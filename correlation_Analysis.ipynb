# correlation Analysis
Correlation Matrix is basically a covariance matrix  It is a matrix in which i-j position defines the correlation between the ith and jth parameter of the given data-set.basically An increase in one variable leads to an increase in the other variable and vice versa
this concludes is we can predicts valuse by fiting line or define one particular variable with the help of other correlated variable 
* There are three types of correlation:
Positive and negative correlation
Linear and non-linear correlation
Simple, multiple, and partial correlation 

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/omgo101/dataset-/main/house-prices.csv")
data
```

```python
data['Price'].describe()
```

**HISTOGRAM**

```python
plt.figure(figsize = (9, 5))
data['Price'].plot(kind ="hist")
```

**Correlation Matrix**

```python
corrmat = data.corr()  
f, ax = plt.subplots(figsize =(9, 8))
sns.heatmap(corrmat, ax = ax, cmap ="YlGnBu", linewidths = 0.1)# linewidths = 0.1 Width of the lines that will divide each cell.
```

```python
import seaborn as sns
 
# checking correlation using heatmap 
#plotting the heatmap for correlation
ax = sns.heatmap(data.corr(), annot=True)
```

**so the highest correlation between price is sqFt**

```python
sns.pairplot(data)
```

```python
sns.relplot(x='Price',y='SqFt',data=data)
```

```python
 #adds the title
plt.title('Correlation')
x=data['Price']
y=data['SqFt']
# plot the data
plt.scatter(x, y)
 
# fits the best fitting line to the data
plt.plot(np.unique(x),
         np.poly1d(np.polyfit(x, y, 1))
         (np.unique(x)), color='red')
 
# Labelling axes
plt.xlabel('Price')
plt.ylabel('SqFt')
```

**but how actually correlation is calculated** 
its called **Pearson Correlation Coefficient** measures the linear association between two variables.
It always takes on a value between -1 and 1 where:

-1 indicates a perfectly negative linear correlation between two variables
0 indicates no linear correlation between two variables
1 indicates a perfectly positive linear correlation between two variables
